```python
import numpy as np
import pandas as pd
import os
from pyts.image import GramianAngularField
from PIL import Image
import cv2
```

```python
threshold = 0.1
output_dir = "gaf_images"
timeframes = ['minute']
```

```python
for label in ['long', 'short', 'neutral']:
    os.makedirs(os.path.join(output_dir, label), exist_ok=True)
output_dir = "gaf_images"
os.makedirs(output_dir, exist_ok=True)
```

```python
minute_data = pd.read_csv('SPY 2023 Data till Sep23.xlsx - Sheet1.csv')
# hourly_data = pd.read_csv('hourly.csv')

minute_close_prices = minute_data['close'].values
# hourly_close_prices = hourly_data['close'].values

# Constants for block size
minute_block_size = 60  
# hourly_block_size = 24 * 2       
```

```python
def create_blocks(data, block_size):
    return np.array([data[i:i + block_size] for i in range(0, len(data) - block_size + 1, block_size)])
```

```python
minute_blocks = create_blocks(minute_close_prices, minute_block_size)
# hourly_blocks = create_blocks(hourly_close_prices, hourly_block_size)
```

```python
gaf = GramianAngularField(method='summation')
```

```python
for i, minute_block in enumerate(minute_blocks[:-1]):
    
    minute_gaf = gaf.transform(minute_block.reshape(1, -1))[0]
    # hourly_gaf = gaf.transform(hourly_block.reshape(1, -1))[0]
    
    
    minute_gaf_img = ((minute_gaf - np.min(minute_gaf)) / (np.max(minute_gaf) - np.min(minute_gaf)) * 255).astype(np.uint8)
    # hourly_gaf_img = ((hourly_gaf - np.min(hourly_gaf)) / (np.max(hourly_gaf) - np.min(hourly_gaf)) * 255).astype(np.uint8)

    pct_change = (minute_blocks[i+1][-1] - minute_blocks[i][-1])/minute_blocks[i][-1]*100

    if(pct_change>threshold):
        label="long"
    elif(pct_change<-threshold):
        label="short"
    else:
        label="neutral"
    
    #saving
    minute_gaf_img_path = os.path.join(output_dir,label,f"minute_gaf_block_{i+1}.png")
    # hourly_gaf_img_path = os.path.join(output_dir, f"hourly_gaf_block_{i+1}.png")
    
    Image.fromarray(minute_gaf_img).save(minute_gaf_img_path)

    cv2.imwrite(minute_gaf_img_path, minute_gaf_img)

    # cv2.imwrite('output_image.png', array)
    
    # Image.fromarray(hourly_gaf_img).save(hourly_gaf_img_path)
    print(f"Minute GAF Matrix Dimensions: {minute_gaf.shape}")
```
